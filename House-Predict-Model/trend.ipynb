```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import pickle
df = pd.read_csv('house_price_cleaning.csv')

# City Convert number

city_label = LabelEncoder()
location_label = LabelEncoder()
df['City'] = city_label.fit_transform(df['City'])
df['Location'] = location_label.fit_transform(df['Location'])


rfr = RandomForestRegressor()
# feature
X = df[['Area_sqft','Bedrooms', 'Bathrooms', 'City','Location']]
y = df['Price']

rfr.fit(X, y)

with open('house_model.pkl','wb') as f:
    pickle.dump(rfr, f)

with open('city_model.pkl','wb') as f:
    pickle.dump(city_label,f)

with open('location_model.pkl','wb') as f:
    pickle.dump(location_label,f)



# check accuracy your model
y_pre = rfr.predict(X)
accuracy = r2_score(y, y_pre)
print(f"R^2 Score: {accuracy}")

# inputing 
areaSqft = int(input("Enter Area in sqft: "))
bedroom = int(input("Enter Number of Bedrooms: "))
bathroom = int(input("Enter Number of Bathrooms: "))
city = input("Enter City (e.g., New York, Los Angeles): ")
loaction = input("Enter Loaction : ")



try:
    encoded_city = city_label.transform([city])[0]
    encode_loaction = location_label.transform([loaction])[0]
 
    new_house = [[areaSqft,bedroom,bathroom,encoded_city,encode_loaction]]

    predect_price = rfr.predict(new_house)[0]

    print(f"Predicted House Price: {city} : location : {loaction} : â‚¹{predect_price:,.2f}")
except ValueError as e:
    print(f"Error: {e}. Please enter a valid city and location from the dataset.")
    exit()

```
