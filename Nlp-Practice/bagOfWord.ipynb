```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Load dataset
df = pd.read_json('ai_roadmap_dataset.json')

model = LogisticRegression()

# Encode answers
l = LabelEncoder()
df['answer_encoded'] = l.fit_transform(df['answer'])

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['question'])
y = df['answer_encoded']

# Use classifier

model.fit(X, y)
model.predict(X)


# Chat loop
try:
    while True:
        user_input = input('You: ')
        if user_input.lower() == 'e':
            print("Bye Bye")
            break

        x_user = vectorizer.transform([user_input])
        pred_label = model.predict(x_user)[0]
        pred_answer = l.inverse_transform([pred_label])[0]
        print(f'Bot: {pred_answer}')

except Exception as e:
    print(e)
```
