```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import torch

# Dataset load
df = pd.read_json('dataset.json')

# Model load
model = SentenceTransformer("all-MiniLM-L6-v2")

# Encode all dataset sentences as tensor
X = model.encode(df['text'].tolist(), convert_to_tensor=True)

while True:
    query = input("You: ")
    if query.lower() == "q":
        print("Bye Bye")
        break

    # Encode user query as tensor
    X_user = model.encode([query], convert_to_tensor=True)

    # Cosine similarity
    similarity = util.cos_sim(X_user, X)  # shape [1, N]
    best_index = similarity.argmax().item()
    best_score = similarity[0][best_index].item()
    # print("Best index:", best_index)

    # Recovery level thresholds
    if best_score <= 0.4:
        print("Recovery level: Low ðŸ˜”")
    elif best_score <= 0.7:
        print("Recovery level: Medium ðŸ™‚")
    else:
        print("Recovery level: High ðŸ˜ƒ")

    # Output best matching sentence and score
    print("Most similar text:", df['text'].iloc[best_index])
    print("Score:", df['score'].iloc[best_index])
    print("Similarity:", round(best_score, 2))
    print("------")
```
