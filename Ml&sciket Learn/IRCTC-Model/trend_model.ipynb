```python
# from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

import pandas as pd
from sklearn.metrics import r2_score

import pickle
# --------------------------
# Step 1: Dummy Dataset
# --------------------------
import pandas as pd

df = pd.read_json("pk.json")

  

# --------------------------
# Step 2: Train Model
# --------------------------
X = df[["SeatsAvailable", "BasePrice"]]
y = df["FinalPrice"]
model = RandomForestRegressor()
model.fit(X, y)

y_pre = model.predict(X)
accuracy = r2_score(y, y_pre)
print(f"R^2 Score: {accuracy}")


keywords_list =['banaya','make','made','develop','create','you']

# keyword 
def keywords_detect(user_input:str):
    user_input = user_input.lower()
    if any(word in user_input for word in keywords_list):
        return 'ask_creator'
    return 'train_price'

while True:
    user_input = input("Ask Any Question For Train Related")
    # exit 

    if user_input.lower() in ["q","e"]:
       print("\nü§ñ Bye Bye Dost")
       break

       # function call pass parameter
    intend = keywords_detect(user_input)

    # print owner
    if intend == 'ask_creator':
        print('I am RailAi Model created by (Pankaj Singh)')

    # train logic 
    elif intend == 'train_price':
       # filtering for train details get one by one 
       train_data = df[df['TrainName'].str.lower().str.contains(user_input.lower())]

        # calculate 
       if train_data is not None and not train_data.empty:
           avg_seats = train_data['SeatsAvailable'].mean()
           base_price = train_data['BasePrice'].iloc[0]
           prediced_price = model.predict([[avg_seats, base_price]])[0]
           explanation = train_data['Explanation'].iloc[0]

           print(f'\n üí∞ Predicted Price For {user_input} : ‚Çπ{prediced_price:.2f}')
        #    print(f" {explanation}")

            # conent find in  array one by one get/find
           for f in explanation:
                 print(f"{f}")
              

       else:
        print(f'‚ùå Train Not Found')

        

    else:
        print("Please Try Again")
```
