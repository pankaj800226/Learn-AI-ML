```python
import pandas as pd

data={
    'name':["aman","sanam","sohan","rawan","rawan"],
    'age':[12,None,23,None,55],
    'salary':[2000,4000,None,None,2500]
}

df = pd.DataFrame(data)
print('original data')
print(df)
df["age"].fillna(df["age"].mean(),inplace=True)
df['salary'].fillna(df["salary"].mean(),inplace=True)
print(df)
print("filter 2200 above")
above_salary = df[df["salary"]>2000]
print(above_salary)


```
