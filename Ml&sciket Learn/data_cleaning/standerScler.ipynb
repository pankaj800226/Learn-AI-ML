Stander Scaler := to use big number to conver small number

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd

data={
    'Hours':[1,3,5,6,9],
    'Score':[100,30,500,40,10]
}

df = pd.DataFrame(data)
staner_scaler = StandardScaler()
sd = staner_scaler.fit_transform(df) 
print('\nStander scler output')

print(pd.DataFrame(sd,columns=[['Hours','Score',]]))

# 1 - 0 number convert
min_max = MinMaxScaler()
min_max_result = min_max.fit_transform(df)
print('\Min Max scler output')

print(pd.DataFrame(min_max_result,columns=[['Hours','Score',]]))

x = df[['Hours']]
y = df[['Score']]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42)

print('Data Traning')
print(x_train)

print('Tranning')
print(x_test)
```
