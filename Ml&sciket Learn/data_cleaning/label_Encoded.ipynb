LabelEncoded use two row select and conver 0 and 1 

eg: male & female
     0       1

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv('simple.csv')

df_label = LabelEncoder()
df['Gender_Encoded'] = df_label.fit_transform(df['Gender'])
df['Passion_Encoded'] = df_label.fit_transform(df['Passion'])

print('\n Final label encoded data')
print(df[['Name','Gender','Gender_Encoded','Passion','Passion_Encoded']])
```
