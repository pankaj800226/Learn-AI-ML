```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# --------------------------
# Step 1: Dataset
# --------------------------
data = {
    "TrainNo": [12301,12301,12302,12302,12303,12303,12304,12304,12305,12305],
    "TrainName": ["Rajdhani Express","Rajdhani Express","Shatabdi Express","Shatabdi Express",
                  "Garib Rath","Garib Rath","Duronto Express","Duronto Express",
                  "Tejas Express","Tejas Express"],
    "SeatsAvailable": [50,10,40,5,60,15,30,8,20,2],
    "BasePrice": [1500,1500,1200,1200,800,800,1400,1400,2000,2000],
    "FinalPrice": [1200,1500,900,1200,600,750,1100,1400,1500,2000]
}

df = pd.DataFrame(data)

# --------------------------
# Step 2: Train Model
# --------------------------
X = df[["SeatsAvailable","BasePrice"]]
y = df["FinalPrice"]

model = LinearRegression()
model.fit(X, y)

# --------------------------
# Step 3: User Input (Only Train Name + DaysBeforeTravel)
# --------------------------
train_name = input("Enter Train Name: ")
# days_before = int(input("Enter days before travel (e.g., 1 for next day): "))

# Filter dataset for that train
train_data = df[df["TrainName"].str.lower() == train_name.lower()]

if not train_data.empty:
    # Take average seats & base price for that train
    avg_seats = train_data["SeatsAvailable"].mean()
    base_price = train_data["BasePrice"].iloc[0]

    # Predict Final Price
    predicted_price = model.predict([[avg_seats, base_price]])[0]

    # --------------------------
    # Extra Rule: 1 din ke liye discount
    # --------------------------
    # if days_before == 1:
    #     predicted_price *= 0.8   # 20% discount
  
    print(f"\nüîπ Predicted Price for {train_name}: ‚Çπ{predicted_price:.2f}")
else:
    print("‚ö†Ô∏è Train not found in dataset!")
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# --------------------------
# Step 1: Dummy Dataset
# --------------------------
data = {
    "TrainNo": [12301,12301,12302,12302],
    "TrainName": ["Rajdhani Express","Bulet Train","Shatabdi Express","Shatabdi Express"],
    "SeatsAvailable": [50,10,40,5],
    "BasePrice": [1500,2500,1200,1200],
    "FinalPrice": [1200,1500,900,1200],
    "Explanation": [
        "Off-peak season discount applied.",
        "High demand, no discount available.",
        "Early bird offer (booked in advance).",
        "Last few seats left, higher price."
    ]
}
df = pd.DataFrame(data)

# --------------------------
# Step 2: Train Model
# --------------------------
X = df[["SeatsAvailable", "BasePrice"]]
y = df["FinalPrice"]
model = LinearRegression()
model.fit(X, y)

# --------------------------
# Step 3:  NLP Intent Detector
# --------------------------
def detect_intent(user_input: str):
    user_input = user_input.lower()
    if any(word in user_input for word in ["banaya", "creator", "develop", "kisne", "made"]):
        return "ask_creator"
    return "train_price"
    

# --------------------------
# Step 4: Handle User Input
# --------------------------

print("ü§ñ Bot: Hi! Type 'e' or 'q' to exit.")


while True:
    user_input = input("\nYou: Ask a question : ")

    if user_input.lower() in ["e", "q"]:
        print("ü§ñ Bot: Bye! Safe journey üöÄ")
        break

    intent = detect_intent(user_input)

    if intent == "ask_creator":
        print("ü§ñ Bot: Ye system Pankaj Kumar ne banaya hai üöÄ")

    elif intent == "train_price":
    # filtering train name and one by one find 
        train_data = df[df["TrainName"].str.lower().str.contains(user_input.lower())]

        if train_data is not None and not train_data.empty:
            avg_seats = train_data["SeatsAvailable"].mean()
            base_price = train_data["BasePrice"].iloc[0]
            predicted_price = model.predict([[avg_seats, base_price]])[0]
            explanation = train_data["Explanation"].iloc[0]

            print(f"ü§ñ Bot: Predicted Price for {user_input} : ‚Çπ{predicted_price:.2f}")
            # print(f"1: {explanation}")
            for i, exp in enumerate(train_data["Explanation"], start=1):
                print(f"  {i}: {exp}") 
        else:
            print("ü§ñ Bot: Sorry, mujhe us train ka data nahi mila ‚ùå")

    else:
        print("ü§ñ Bot: Mujhe samajh nahi aaya, kya aap dubara try karenge?")
```
