MAE MSE RMSE


```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np

real_score = [90,60,80,105]

predict_score = [85,70,70,95]

mae = mean_absolute_error(real_score,predict_score)

mse = mean_squared_error(real_score,predict_score)

rmse = np.sqrt(mse)

print("mae avarage value : ",mae)
print("mse mistak square value : ",mse)
print("rmse final realistic error : ",rmse)
```
